```python
df = pd.read_csv(csv_paths[-2])
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
# filter out if all null values assigned under all group_*** columns
cond = (df['group_ORG'].isna()) & (df['group_PER'].isna()) & (df['group_MISC'].isna()) & (df['group_title_a'].isna())

new_df = df[~cond]
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
group_lst = ['group_PER', 'group_ORG', 'group_MISC', 'group_title_a']
dict(new_df.loc[0, group_lst]).values()
```

```python
# filter values that have duplicate of their own or just a single string
for ind in new_df.index:
  val_lst = list(dict(new_df.loc[ind, group_lst]).values())
  string_only = [val for val in val_lst if isinstance(val, str)]

  if len(string_only) > 1 and string_only[0] == string_only[1]:
    new_df.at[ind, 'final_group'] = string_only[0]
  elif len(string_only) == 1:
    new_df.at[ind, 'final_group'] = string_only[0]

new_df
```
